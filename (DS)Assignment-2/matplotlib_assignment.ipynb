```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [10,15,25,30,50]
```

```python
plt.plot(x,y,marker = 'o',linestyle='-',color='g') #Decoration

plt.title("Total distance covered in (KM)")
plt.xlabel("Days")
plt.ylabel("Distance(KM)")
plt.show()
```

## üéìTask 2:

```python
students = ['john','Jane','Alice','Bob']
marks = [75,85,60,90]
plt.bar(students,marks,color=['yellow','orange','red','green']) #Decoration---> yellow = Not Bad
                                                                                #orange = Good
                                                                                #red = Average
                                                                                #green = Excellent

plt.title("Marks Obtained by Each Student")
plt.xlabel("Students Name")
plt.ylabel("Marks obtained")
plt.show()
```

## üéìTask 3:

```python
regions = ['North America','Europe','Asia','Others']
percentage = [45,25,20,10]
colors = ['green','orange','yellow','red']

plt.pie(percentage,labels=regions,autopct='%1.1f%%',colors=colors)
plt.title('Distribution of Revenue from Different Regions')
plt.show()

#Green--> Highest Revenue
#orange--> Lesser revenue than green
#yellow--> lesser than orange
#red--> Lowest Revenue
```
