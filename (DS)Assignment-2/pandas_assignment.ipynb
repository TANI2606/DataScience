```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'Name':['Alice','Bob','Charlie','Diana','Eve'],
    'Age':[24,27,22,32,29],
    'Department':['HR','Finance','IT','Marketing','HR'],
    'Salary':[45000,54000,50000,62000,47000]
})
df.head(5)
```

```python
df.to_csv("job.csv", index = False)
```

```python
df['Age'].describe()
```

```python
df['Salary'].describe()
```

```python
HR_salary = df[df['Department'] == 'HR']['Salary'].mean()
print('HR Avg salary will be: ',HR_salary)
```
