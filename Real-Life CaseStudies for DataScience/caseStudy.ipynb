## ðŸ‘‘Master:Case study 1:(Bank Customer Transaction Mangement Using Function)

```python
#step1:Make an empty dict.

customers = {}
```

```python
#step2:Add new customers 
def add_customers(account_name,name,balance):
    if account_name in customers:
        print("Account already existed!")
    else:
        customers[account_name]={"Name":name,"Balance":balance}
        print("Customer added successfully!")
```

```python
#Adding customers
add_customers(101,"Akshat",800000)
```

```python
#step4:Function to display customer details 
def display_customer(account_name):

    if account_name in customers:
        print(f"Account no: {account_name}")
        print(f"Name: {customers[account_name]['Name']}")
        print(f"Balance: â‚¹{customers[account_name]['Balance']} ")
    else:
        print("Customers Not found!")
```

```python
#step5:Displaying customer
display_customer(101)
```

```python
#step6:Updating the customer balance
def update_balance(account_name,amount,transaction_type):
    if account_name in customers:
        if transaction_type == "deposit":
            customers[account_name]["Balance"] += amount
        elif transaction_type == "withdraw" and customers[account_name]["Balance"] >= amount:
            customers[account_name]["Balance"] -= amount
        else:
            print("Insuffient Balance!")
        print("Transaction sucessfully!")
    else:
        print("Account not Found!")
        
```

```python
#step7:Updatind balances
update_balance(101,780000,"withdraw")
display_customer(101)
```

## ðŸ’¡Case study 8:(Insurance claim Records using Numpy.)

```python
import numpy as np

claims = np.array([
    [12000,14500,18000], #policyholder1
    [16700,12450,17350], #policyholder2
    [7500,11000,10050]   #policyholder3
])
print("Insurance Claim Data (INR):")
print(claims)
```

```python
print("Claims amount for Policyholder 3 in Month 1:",claims[2,0])
print("Claims amount for Policyholder 1 in Month 3:",claims[0,2])
```

```python
#Total claim Amount per Policyholder
total_claims = np.sum(claims,axis=1)
print("Total claims per Policyholder:",total_claims)
```

```python
#Average amount claim per month
avg_claims = np.mean(claims,axis=0)
print("Average amount claim per month:",avg_claims)
```

```python
#Finding claims greater than 15000
high_value_claim = claims[claims > 15000]
print("High Value claims:",high_value_claim)
```

## ðŸ‘‘Master:Case study 2:(Customer Feedback Analysis in Retail Industry using Numpy)

```python
import numpy as np

#Step1:simulate importing mixed dataset (numbers & text)
#each row represents: Customer ID,Rating(1-5) and feedback comment


data = np.array([
    ["C001",5,"Excellent service and friendly staff"],
    ["C002",4,"Good product selection but long wait time"],
    ["C003",3,"Average experience,checkout was slow"],
    ["C004",5,"Loved the discount"],
    ["C005",2,"Poor customer support"],
    ["C006",4,"Good product but expensive"],
    ["C007",1,"Terrible experience"],
    ["C008",5,"Amazing deals"],
])
```

```python
print(data)
```

```python
#Step2:extract numerical rating for analysis
rating = data[:, 1].astype(int)
print(f"Total feedback entries: {len(rating)}")
```

```python
#Step3:performing statistical analysis
avg_ratings = np.mean(rating)
print(f"Average customer rating: {avg_ratings:.2f}")
```

```python
highest_ratings = np.max(rating)
print(f"Highest rating recieved: {highest_ratings}")

lowest_ratings = np.min(rating)
print(f"Lowest rating recieved: {lowest_ratings}")
```

```python
unique_ratings,rating_counts = np.unique(rating,return_counts=True)
```

```python
#Step4:extract text feedback for qualitive insights
feeback_comments = data[:, 2]

#Step5:display insights
print("\n**Customer Feedback Analysis Summary**")
print()

print("\n**Rating Distribution:**")
print()
for rating,count in zip(unique_ratings,rating_counts):
    print(f"Rating {rating}: {count} customers")
```

```python
print("\n**Sample Customer Feedback Comments:**")
print()

for comment in feeback_comments[:5]:
    print(f"- {comment}")
```
