# Matplotlib-> Data Visualization basics:

## Example 1:(Graph)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
ax.set_title("Basic Figure And Axes")
ax.set_xlabel("Days")
ax.set_ylabel("total patients")
plt.show()
```

## Example 2:

```python
months = ['Jan','Feb','Mar','Apr','May','Jun']
recovery_rate = [52,70,23,79,65,96]
plt.plot(months,recovery_rate,marker = 'o',linestyle='-',color='g',label='Recovery Rate') #Decoration

plt.title("Patient Recovery Trends")
plt.xlabel("Months")
plt.ylabel("Recovery Percentage")
plt.legend()  #Recovery rate button
plt.show()
```

## Example 3:(Bar Plot)

```python
medicines = ['Paracetamol','Antibiotics','Painkillers','Insulin','Antacids']
demand = [550,450,615,575,525]
plt.bar(medicines,demand,color=['green','red','gold','green','green']) #Decoration

plt.title("Medicines Demand Analysis")
plt.xlabel("Medicine Type")
plt.ylabel("Units sold")
plt.show()
```

# Example 4:(Pie Chart)

```python
diseases = ['Diabetes','HyperTension','Heart Diseases','Asthma','Others']
percentage = [30,25,15,10,20]
colors = ['blue','pink','purple','orange','red']

plt.pie(percentage,labels=diseases,autopct='%1.1f%%',colors=colors)
plt.title('Diseases Distribution in City')
plt.show()
```

# Example 5:(Scatter Plot)

```python
time_spent = [10,15,20,25,30,35,40]
diagonosis_accuracy = [60,65,70,75,80,85,90]

plt.scatter(time_spent,diagonosis_accuracy,color='red')
plt.xlabel('Time Spent by Doc(minutes)')
plt.ylabel;('Diagnosis Accuracy (%)')
plt.title('Doctor-Patient Interaction Time vs. Diagnosis Accuracy')
plt.show()
```

# Example 6:(Histogram)

```python
ages = np.random.randint(10,80,100)

plt.hist(ages,bins=10,color='grey',edgecolor='black')
plt.xlabel('Age Groups')
plt.ylabel;('Number of Patient')
plt.title('Age Distribution of Patients')
plt.show()
```

# Example 7:(Word Cloud)

```python
from wordcloud import WordCloud
text = 'fever cough cold headache nausea fatigue pain weakness dizziness allergy infection sore throat congestion'
wordcloud = WordCloud(width=800,height=400,background_color='white').generate(text)

plt.figure(figsize=(10,5))
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis("off")
plt.title('Common Symtoms Analysis')
plt.show()
```

# Seaborn->Another Data Visualization Lib:

## Example 1:(Scatterplot)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("telecom_customer_data.csv")
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df,x='Call_Duration',y='Satisfaction_Score',hue='Network_Provider')

plt.legend(title='Network Provider',bbox_to_anchor=(1,1),loc = 'upper left')

plt.xlabel('Call Duration (minutes)')
plt.ylabel;('Satisfaction Score')
plt.title('Call duration vs Satisfaction score')
plt.show()
```

## Example 2:(CategorialPlots)

```python
plt.figure(figsize=(6,5))
sns.countplot(data=df,x='Network_Provider',hue='Network_Provider',palette='Set2')

plt.xlabel('Network Provider')
plt.ylabel;('Count of customers')
plt.title('Customer Distribution by Network Provider')

plt.show()
```

## Example :(DistributionPlots)

```python
plt.figure(figsize=(7,4))
sns.histplot(data=df['Call_Duration'],bins=20,kde=True,color='purple')

plt.xlabel("Call Duration(minutes)")
plt.ylabel;("Frequency")
plt.title("Distribution of Call Duration")

plt.show()
```

## Example 4:(MatrixPlots)

```python
plt.figure(figsize=(8,5))
numeric_df = df.select_dtypes(include=['number'])
sns.heatmap(numeric_df.corr(),annot=True,cmap='coolwarm',fmt='.2f',linewidths=0.5)

plt.title("Correlation Matrix of Features")
plt.show()
```

## Example 5:(BoxPlots)

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='Network_Provider',y='Satisfaction_Score')

plt.xlabel("Network Provider")
plt.ylabel;("Satisfaction Score")
plt.title("Satisfaction Score Across Different Network Providers")

plt.show()
```

## Example 6:(ViolinPlots)

```python
plt.figure(figsize=(8,5))
sns.violinplot(data=df, x='Network_Provider',y='Satisfaction_Score')

plt.xlabel("Network Provider")
plt.ylabel;("Satisfaction Score")
plt.title("Satisfaction Score Across Different Network Providers")

plt.show()
```
