## Numpay Basics:

```python
import numpy as np
```

```python
#2-D array
arr2 = np.array([[1,2,3], [4,5,6]])
print("2D array:\n",arr2)
```

```python
#sms2:
arr_a = np.array([1,2,3])
arr_b = np.array([4,8,56])

print("The addition:",arr_a+arr_b)
print("The multiplication:",arr_a*arr_b)
print("The square of arr_a:",arr_a**2)
print("The square of arr_b:",arr_b**2)
```

```python
#Reshaping
reshaped = np.arange(12).reshape(3,4)
print("Reshaped(3X4):\n\n",reshaped)
```

## Numpy Useful Functions:

```python
arr = np.array([23,5,56,11,1,32])
print("Mean:",np.mean(arr))
```

```python
print("Standard deviation:",np.std(arr))
```

```python
print("Sorted:",np.sort(arr))
```

```python
import numpy as np
mat = np.arange(1,26).reshape(5,5)
print("(5X5):\n\n",mat)
```

```python
sum_mat = np.sum(mat,axis = 1)
print(sum_mat)
```

```python
sum_mat = np.sum(mat,axis = 0)
print(sum_mat)
```
