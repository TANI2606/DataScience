## How to Find Missing Values

```python
cols = df.columns
cols
```

```python
import seaborn as sns
cols = df.columns
colours = ['#000099','#ffff00']
sns.heatmap(df[cols].isnull(),cmap=sns.color_palette(colours))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

#To check outliers,if there is outlier use (median)
plt.figure(figsize=(8,4))
sns.boxplot(y=df['Wind'])
plt.show()
```

```python
#To check outliers by statistical method(Best method)
Q1 = df['Wind'].quantile(0.25)
Q3 = df['Wind'].quantile(0.75)
IQR = Q3-Q1

#Define outlier bounds
lower_bound = Q1-1.5*IQR
upper_bound = Q3+1.5*IQR

#Find Outlier
outliers = df[(df['Wind']< lower_bound) | (df['Wind']> upper_bound)]
print("Outliers: ",outliers.shape[0])
```
